```python
# 导入 keras 应用模块
from keras.applications.inception_v3 import InceptionV3
from EnLIMELateestImage import EnLIMEExplain
# 生成模型实例
inceptionV3_model = InceptionV3()

# 创建 LIMEExplain 实例
lime_explainer = EnLIMEExplain(inceptionV3_model, num_samples=20, perturb_ratio=[0.2, 0.8],
                               regressor_type='linear', weight_type='kernel_shap', perturbation_type='fix')
# 对特定的图像进行解释
importances = lime_explainer.explain_instance("C://Users//User//Desktop//dogexample.png")

# 打印重要性得分
print(importances)
```
